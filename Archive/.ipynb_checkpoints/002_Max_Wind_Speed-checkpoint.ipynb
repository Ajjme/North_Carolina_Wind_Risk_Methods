```python
from netCDF4 import Dataset
import numpy as np

# File paths
input_file = 'wind_speed_merged197910.nc'  # Adjust path as needed
output_file = 'max_wind_speed.nc'

# Open the input NetCDF file
with Dataset(input_file, 'r') as src:
    # Read dimensions
    south_north = src.dimensions['south_north'].size
    west_east = src.dimensions['west_east'].size

    # Read variables
    lats = src.variables['XLAT'][:]
    lons = src.variables['XLONG'][:]
    wind_speed = src.variables['wind_speed'][:]

    # Compute maximum wind speed along the time dimension
    max_wind_speed = np.nanmax(wind_speed, axis=0)

# Create a new NetCDF file
with Dataset(output_file, 'w', format='NETCDF4') as dst:
    # Define dimensions
    dst.createDimension('south_north', south_north)
    dst.createDimension('west_east', west_east)

    # Define variables
    lat_var = dst.createVariable('XLAT', 'f4', ('south_north', 'west_east'))
    lon_var = dst.createVariable('XLONG', 'f4', ('south_north', 'west_east'))
    max_wind_var = dst.createVariable('max_wind_speed', 'f4', ('south_north', 'west_east'))
    
    # Add attributes
    lat_var.units = 'degree_north'
    lon_var.units = 'degree_east'
    max_wind_var.units = 'm s-1'
    max_wind_var.long_name = 'Maximum wind speed at 10 meters'
    
    # Write data to variables
    lat_var[:] = lats
    lon_var[:] = lons
    max_wind_var[:] = max_wind_speed

print(f"Max wind speed file created: {output_file}")
```

```python
import numpy as np
import matplotlib.pyplot as plt
from netCDF4 import Dataset
import cartopy.crs as ccrs
import cartopy.feature as cfeature

# File path
input_file = 'max_wind_speed.nc'  # Adjust path as needed

# Open the NetCDF file
with Dataset(input_file, 'r') as src:
    lats = src.variables['XLAT'][:]
    lons = src.variables['XLONG'][:]
    max_wind_speed = src.variables['max_wind_speed'][:]

# Plot the data
plt.figure(figsize=(12, 8))
ax = plt.axes(projection=ccrs.PlateCarree())

# Add features to the map
ax.add_feature(cfeature.COASTLINE)
ax.add_feature(cfeature.BORDERS, linestyle=':')
ax.add_feature(cfeature.LAND, edgecolor='black')
ax.add_feature(cfeature.OCEAN, facecolor='lightblue')

# Plot the wind speed as a color map
c = ax.pcolormesh(lons, lats, max_wind_speed, cmap='viridis', transform=ccrs.PlateCarree())
plt.colorbar(c, ax=ax, orientation='horizontal', pad=0.05, label='Max Wind Speed (m/s)')

# Add labels and title
plt.title('Maximum Wind Speed at 10 Meters', fontsize=16)
ax.set_xlabel('Longitude')
ax.set_ylabel('Latitude')

# Show the plot
plt.show()
```

```python
import xarray as xr
import pandas as pd
import numpy as np

# Open the NetCDF file using xarray
ds = xr.open_dataset('wind_speed_merged197910.nc')

# Extract the necessary variables from the dataset
lat = ds['XLAT'].values
lon = ds['XLONG'].values
wind_speed = ds['wind_speed'].values  # Shape is (time, south_north, west_east)

# Flatten the lat, lon arrays to match the wind speed
lat_flat = lat.flatten()
lon_flat = lon.flatten()

# Initialize an empty list to store the max wind speed for each lat-lon combination
max_wind_speeds = []

# Loop through each (lat, lon) pair and find the maximum wind speed over time
for i in range(lat.size):
    for j in range(lon.size):
        # Extract the wind speed for each lat-lon pair across time
        wind_speed_values = wind_speed[:, i, j]
        
        # Calculate the maximum wind speed, ignoring NaN values
        max_wind = np.nanmax(wind_speed_values)
        
        # Append the results to the list
        max_wind_speeds.append([lat_flat[i], lon_flat[j], max_wind])

# Create a DataFrame
df = pd.DataFrame(max_wind_speeds, columns=['Latitude', 'Longitude', 'Max Wind Speed (m/s)'])

# Show the resulting DataFrame
print(df)
```

```python
import geopandas as gpd
df_counties = gpd.read_file("/data/HW/HW2/NCDOT_County_Boundaries.shp")
```

```python
df_counties['geomlist'] = df_counties['geometry'].apply(lambda x: list(x.exterior.coords))
```

```python
def point_inside_polygon(lat,lng,poly):
    n = len(poly)
    inside =False
    p1x,p1y = poly[0]
    for i in range(n+1):
        p2x,p2y = poly[i % n]
        if lat > min(p1y,p2y):
            if lat <= max(p1y,p2y):
                if lng <= max(p1x,p2x):
                    if p1y != p2y:
                        xinters = (lat-p1y)*(p2x-p1x)/(p2y-p1y)+p1x
                    if p1x == p2x or lng <= xinters:
                        inside = not inside
        p1x,p1y = p2x,p2y
    return inside
```

```python
poly = df_counties.loc[0,'geomlist']
lat=36
lng= -75
point_inside_polygon(lat, lng, poly)
```

```python
def get_counties(row):
    for ix, area in df_counties.iterrows():
        is_in_area=False
        if row['latitude'] and row['longitude']:
            is_in_area = point_inside_polygon(row['latitude'], row['longitude'], area['geomlist'])
            if is_in_area:
                #found area, exit
                return area['CountyName']
    return ""
df['counties'] = df.apply(lambda row: get_counties(row), axis=1)
```
