```python
import netCDF4 as nc

# File path
file = '/data/era-interim/era_interim_monthly_197901_201512.nc'

# Open the NetCDF file
fp = nc.Dataset(file)

# Print the global attributes
print("Global Attributes:")
for attr in fp.ncattrs():
    print(f"{attr}: {getattr(fp, attr)}")

# Check for projection information in variables
print("\nCoordinate Variables and Their Attributes:")
for var in fp.variables:
    print(f"\nVariable: {var}")
    for attr in fp.variables[var].ncattrs():
        print(f"  {attr}: {getattr(fp.variables[var], attr)}")

fp.close()
```
