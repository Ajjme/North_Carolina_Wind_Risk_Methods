```python
import xarray as xr
import numpy as np
import pandas as pd

# Open the NetCDF file using xarray
ds = xr.open_dataset("/shared_space/ACV/extracted_raw/merged_files/merged197910.nc")

# Extract variables
latitudes = ds['XLAT'].values
longitudes = ds['XLONG'].values
u10 = ds['U10'].values
v10 = ds['V10'].values

# Calculate the wind speed (m/s)
wind_speed = np.sqrt(u10**2 + v10**2)

# Find the maximum wind speed across all time steps
max_wind_speed = np.nanmax(wind_speed, axis=0)

# Flatten the arrays
latitudes_flat = latitudes.flatten()
longitudes_flat = longitudes.flatten()
max_wind_speed_flat = max_wind_speed.flatten()

# Create a DataFrame
df = pd.DataFrame({
    'Latitude': latitudes_flat,
    'Longitude': longitudes_flat,
    'Max Wind Speed (m/s)': max_wind_speed_flat
})
print(df)
```

```python
import geopandas as gpd
import pandas as pd
from shapely.geometry import Point

gdf = gpd.GeoDataFrame(
    df, geometry=gpd.points_from_xy(df.Longitude, df.Latitude), crs="EPSG:4326"
)
# Convert df to GeoDataFrame
#df['geometry'] = df.apply(lambda row: Point(row['Longitude'], row['Latitude']), axis=1)
#points = gpd.GeoDataFrame(df, geometry='geometry', crs="EPSG:4326")

# Perform spatial join
#points_with_counties = gpd.sjoin(points, counties, how="left")

# Select relevant columns
#points_with_counties = points_with_counties[['Latitude', 'Longitude', 'Max Wind Speed (m/s)', 'CountyName']]

print(gdf)
```

```python
import xarray as xr
import matplotlib.pyplot as plt
import cartopy.crs as ccrs
import cartopy.feature as cfeature

# Load the NetCDF file
file_path = '/data/era-interim/era_interim_monthly_197901_201512.nc'
ds = xr.open_dataset(file_path)

# Print dataset summary
print(ds)

# Select a variable to plot (e.g., 't_skin' - skin temperature)
variable = 't_skin'
time_index = 0  # Index for the first time step

# Extract data for the selected time
data = ds[variable].isel(time=time_index)

# Set up the map projection
projection = ccrs.PlateCarree()
fig, ax = plt.subplots(subplot_kw={'projection': projection}, figsize=(12, 6))

# Plot the data
data.plot(ax=ax, transform=ccrs.PlateCarree(), cmap='coolwarm', cbar_kwargs={'label': ds[variable].units})

# Add map features
ax.add_feature(cfeature.COASTLINE)
ax.add_feature(cfeature.BORDERS, linestyle=':')
ax.add_feature(cfeature.LAND, edgecolor='black')
ax.set_title(f"{variable} at time index {time_index}")

# Show the plot
plt.show()
```

```python
import xarray as xr
import matplotlib.pyplot as plt
import cartopy.crs as ccrs
import cartopy.feature as cfeature
import geopandas as gpd
import rasterio
from rasterio.mask import mask
import numpy as np

# Load the NetCDF file
file_path = '/data/era-interim/era_interim_monthly_197901_201512.nc'
ds = xr.open_dataset(file_path)

# Print dataset summary
print(ds)

# Select a variable to plot (e.g., 't_skin' - skin temperature)
variable = 't_skin'
time_index = 0  # Index for the first time step

# Extract data for the selected time
data = ds[variable].isel(time=time_index)

# Load the shapefile
shapefile_path = "/data/HW/HW2/NCDOT_County_Boundaries.shp"
gdf = gpd.read_file(shapefile_path)

# Convert the shapefile to the same CRS as the dataset (PlateCarree projection)
gdf = gdf.to_crs(epsg=4326)  # This assumes the dataset uses WGS84 (EPSG:4326)

# Convert the shapefile to a format that can be used for clipping (GeoJSON format)
geometry = gdf.geometry.values[0]

# Clip the data (apply the mask)
# Convert the NetCDF data into a 2D numpy array and get the latitude and longitude
lat = ds['latitude'].values
lon = ds['longitude'].values

# Create a meshgrid for lat and lon
lon_grid, lat_grid = np.meshgrid(lon, lat)

# Mask the data using the shapefile boundary
# Create a mask based on the shapefile boundary
from shapely.geometry import mapping

# Convert the shapefile geometry to a format that can be used by rasterio
shapefile_geometry = [mapping(geometry)]

# Clip the data using rasterio
masked_data, _ = mask(rasterio.open(data), shapefile_geometry, crop=True)

# Plot the clipped data
projection = ccrs.PlateCarree()
fig, ax = plt.subplots(subplot_kw={'projection': projection}, figsize=(12, 6))

# Plot the clipped data
plt.contourf(lon, lat, masked_data, cmap='coolwarm')
plt.colorbar(label=ds[variable].units)

# Plot the shapefile boundary
gdf.plot(ax=ax, facecolor='none', edgecolor='black', linewidth=0.5)

# Add map features
ax.add_feature(cfeature.COASTLINE)
ax.add_feature(cfeature.BORDERS, linestyle=':')
ax.add_feature(cfeature.LAND, edgecolor='black')
ax.set_title(f"{variable} at time index {time_index}")

# Show the plot
plt.show()

```
