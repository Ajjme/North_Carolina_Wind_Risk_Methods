```python
import numpy as np
import xarray as xr
import matplotlib.pyplot as plt
import glob
import cartopy.crs as ccrs
import cartopy.feature as cfeature
```

```python
class map_generator:
    def __init__(self, list_files):
        self.file_names=list_files
        
    def agg_max(self,year):
        map_files=[file for file in self.file_names if file[-9:-5]==year]

        for i,file in enumerate(map_files):
            fp=xr.open_dataset(file)
            #max_mag_index=fp.max(dim=['south_north','west_east']).argmax()['mag']
    
            if i==0:
                #max_mag=fp.isel(Time=max_mag_index)
                max_mag=fp.max(dim=['Time'])
            else:
                #max_mag_month=fp.isel(Time=max_mag_index)
                max_mag_month=fp.max(dim=['Time'])
                max_mag=xr.concat([max_mag,max_mag_month],dim='Time')
                

        max_value=max_mag.max(dim=['Time'])['mag']
        return max_value
    
    def max_plotter(self,year):
        dem=self.agg_max(year)
        latitudes = dem.coords['south_north']  # Adjust this based on your data
        longitudes = dem.coords['west_east'] 
        extent=[longitudes.min(), longitudes.max(), latitudes.min(), latitudes.max()]
        fig,ax=plt.subplots(figsize=(12, 10), subplot_kw={'projection': ccrs.PlateCarree()})
        c = ax.imshow(dem, origin='lower', cmap='viridis',extent=extent)
        title_string='%s Max Wind Speed (m/s)'
        ax.set_title(title_string%year)
        ax.set_xticks([])
        ax.set_yticks([])
        cbar=fig.colorbar(c,ax=ax,orientation='vertical',shrink=0.6)
        plt.show()   
```

```python
## can change this here. 2022 is the most recent year but you can go back 
year_of_interest='1980'

cdf_list=glob.glob('/shared_space/ACV/extracted_raw/mag_dir/*')
yealy_max=map_generator(cdf_list)
yealy_max.max_plotter(year_of_interest)
```
